    📘 Chapter 1: Quantum Fourier Transform (QFT)

</h2>

<div style="font-family: 'Times New Roman', Times, sans-serif; line-height:1.6; font-size:18px">

<p>
The <b>Quantum Fourier Transform (QFT)</b> is the quantum analogue of the <b>Discrete Fourier Transform (DFT)</b>. It is a fundamental tool in quantum computing
and plays a central role in algorithms such as <b>Quantum Phase Estimation</b> and <b>Shor’s factoring algorithm</b>.
</p>

<h4 style="color:blue;font-weight:bold">🔹 Classical Background: Discrete Fourier Transform (DFT)</h4>

<p>
The <b>Discrete Fourier Transform (DFT)</b> is a fundamental mathematical tool used to analyze finite sets of data by decomposing them into their underlying frequency components. It provides a bridge between two complementary representations of information — typically referred to as the <i>time (or spatial) domain</i> and the <i>frequency domain</i>.
</p>

<p>
The DFT takes a sequence of complex numbers: $(x_0,x_1,x_2, \dots,x_{N-1})$ and transforms it into another sequence: $(y_0,y_1, y_2, \dots, y_{N-1})$
</p>

<p> Where,
$${\large y_k = \frac{1}{\sqrt{N}}\sum_{j=0}^{N-1}x_j\omega_N^{jk}}=\frac{1}{\sqrt{N}}\sum_{j=0}^{N-1} x_j \exp\left(i 2\pi\frac{jk}{N}\right)$$
</p>

<h4 style="color:#1B4F72;">🔹 Quantum Fourier Transform (QFT)</h4>

<p>
For a quantum system of <b>$m$ qubits</b> (with <b>$N = 2^m$</b> basis states), the QFT acts on a computational basis state $|x\rangle$ as:     
</p>

$$
|\phi_y\rangle = \frac{1}{\sqrt{N}}\sum_{x=0}^{N-1}\omega_N^{yx} |x\rangle = \frac{1}{\sqrt{N}}\sum_{x=0}^{N-1} {\large e^{\frac{2\pi iyx}{N}}} |x\rangle
$$

<p>
This transformation maps the state from the <b>computational basis</b> to the <b>Fourier basis</b>, where each output state is a superposition of all basis states,
with information encoded in their <b>relative phases</b>.
</p>

<p>
Importantly, the QFT is a <b>unitary and reversible transformation</b>, ensuring that no information is lost. Unlike the classical DFT, which requires
$O(N^2)$ operations, the QFT can be implemented efficiently on a quantum computer using a sequence of <b>Hadamard</b> and <b>controlled phase rotation gates</b>,
achieving a complexity of $O((log N)^2)$.
</p>

<p>
This exponential improvement makes QFT a powerful computational primitive, enabling quantum algorithms to outperform their classical counterparts in problems involving
periodicity, phase estimation, and number-theoretic structures.
</p>

</div>


<h3 style="color:blue; font-family:'Times New Roman', serif; font-weight:bold">
    🔍 Qiskit Implementation
</h3>

```python
# Prepareing General QFT Circuit
def qft(n):
    qc = QuantumCircuit(n) # n qubit Quantum Circuit
    for j in range(n):
        qc.h(j)
        for k in range(j+1, n):
            qc.cp(np.pi / 2**(k-j), k, j)
    # swap qubits
    for i in range(n//2):
        qc.swap(i, n-i-1)

    return qc
```

```python
# Use QFT Circuit
qc = QuantumCircuit(3)
qc.x(0)
qc.barrier()
qc.compose(qft(2), inplace=True) # inplace comand used to modify your circuit
qc.draw("mpl")
```

```python
# Generic function for decimal input
def qft_decimal(x, n_qubits=3,qft_n = 3):
    if qft_n >n_qubits:
        qft_n = n_qubits

    qc = QuantumCircuit(n_qubits, n_qubits)
    # convert decimal to binary
    binary = format(x, f'0{n_qubits}b')
    # prepare input state
    for i, bit in enumerate(reversed(binary)):
        if bit == '1':
            qc.x(i)
    qc.barrier()

    # apply QFT
    qc.compose(qft(qft_n), inplace=True)
    qc.barrier()
    psi = Statevector.from_instruction(qc)
    # perform measurements
    qc.measure(range(n_qubits), range(n_qubits))

    # Run Simulator
    simulator = AerSimulator()
    tqc = transpile(qc, simulator)

    result = simulator.run(tqc, shots=1024).result()
    counts = result.get_counts()
    return psi, qc, counts
```

```python
psi, qc, counts = qft_decimal(1,2,2)  # x= integer, n_qubits = no of qubits of circuit, qft_n = order of qft
qc.draw("mpl")
```

```python
plot_histogram(counts, title="Measurement After QFT")
```
