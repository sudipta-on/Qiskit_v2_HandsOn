        From Classical Bits to Quantum Bits</summary>

<div style="font-family: sans-serif; line-height:1.55; color:#1e293b;">
  
  <p style="margin:6px 0 14px 0;">
    In this section, we define <b>quantum bits (qubits)</b> and present the mathematical tools used to describe quantum states.
    We begin with classical bits, then transition to their quantum counterparts.  
    For detailed explanations, see <a href="https://profmcruz.wordpress.com/wp-content/uploads/2017/08/quantum-computation-and-quantum-information-nielsen-chuang.pdf" style="color:blue; font-weight:bold">Nielsen &amp; Chuang (2011)</a>.
  </p>

  <!-- Classical Bits -->
  <div style="background:#ffffff; border:1px solid #cbd5e1; border-radius:10px; box-shadow:0 4px 10px rgba(0,0,0,.08); margin-bottom:16px;">
    <div style="background:linear-gradient(135deg,#60a5fa,#2563eb); color:white;  padding:8px 12px;">
      <h3 style="color:white; font-weight:bold; font-size:18px;">🖥 Classical Bits</h3>
    </div>
    <div style="padding:10px 14px; font-size:14px; color:#334155;">
      <p>
      In classical computing, a <b>bit</b> is the basic unit of information. A bit can only exist in one of two distinct states:
                                          $$ 0 \quad \text{or} \quad 1 $$
    </p>
    <p>
      Each bit holds a definite value at any given time. When multiple bits are combined,
      they can represent more complex information.  <br>
      For example, three classical bits can represent
      $ \qquad 2^3 = 8 \quad \text{states: } 000, 001, \dots, 111.$
    </p>
    </div>
  </div>

  <!-- Quantum Bits -->
  <div style="background:#ffffff; border:1px solid #cbd5e1; border-radius:10px; box-shadow:0 4px 10px rgba(0,0,0,.08);">
    <div style="background:linear-gradient(135deg,#34d399,#059669); padding:8px 12px;">
      <h3 style="color:white; font-weight:bold; font-size:18px;">⚛ Quantum Bits (Qubits)</h3>
    </div>
    <div style="padding:10px 14px; font-size:14px; color:#334155;">
    <p>
      Analogous to the role of a bit in classical computation, the fundamental element of
      quantum computation is the <b>quantum bit (qubit)</b>. Unlike a classical bit, a qubit
      can exist in a <b>superposition</b> of states.
    </p>  
    <!-- Single Qubit -->
    <h4 style="color:green; font-weight:bold; font-size:18px; font-family:'Times New Roman', Times, serif; margin-top:10px;">
        Single-Qubit State</h4>
    <p>
      A single qubit is represented as a unit vector in a two-dimensional complex vector space:
    </p>
    <p style="text-align:center; margin:8px 0;">
      $$|a\rangle =
      \begin{bmatrix} a_1 \\ a_2 \end{bmatrix},
      \quad |a_1|^2 + |a_2|^2 = 1$$
    </p>
    <p>
      In Dirac notation (<b>ket</b> Notation):
        $$|a\rangle = a_1|0\rangle + a_2|1\rangle$$
      where the computational basis states are
      $$|0\rangle = \begin{bmatrix}1 \\ 0\end{bmatrix}, \qquad
      |1\rangle = \begin{bmatrix}0 \\ 1\end{bmatrix}.$$
    </p>
    <p style="margin:10px 0; padding:8px 12px; background:#f0fdf4; border-left:4px solid #16a34a; font-size:13px; color:#14532d;">
      📌 <b>Measurement:</b> Upon measurement, outcome <b>0</b> occurs with probability \(|a_1|^2\),
      and outcome <b>1</b> with probability \(|a_2|^2\).  
      The normalization ensures that probabilities sum to one.
    </p>
    <p>
      The conjugate transpose of $|a\rangle$ is called its <b>bra</b>, denoted:
    $$\langle a| = a_1^* \langle 0| + a_2^* \langle 1|,
      \quad \langle 0| = [1, 0], \; \langle 1| = [0, 1]$$
    </p>
    <p style="margin:10px 0; padding:8px 12px; background:#eef2ff; border-left:4px solid #4338ca; font-size:13px; color:#312e81;">
      ✨ <b>Remark:</b> Unlike classical bits that are deterministic, qubits allow
      <b>superposition</b>, a cornerstone of quantum computational power.
    </p><br>
    <!-- Two Qubits -->
  <h4 style="color:green; font-weight:bold; font-size:18px; font-family:'Times New Roman', Times, serif; margin-top:10px;">
        Two-Qubit State</h4>
  <div style="background:#ffffff; border:1px solid #fecaca; border-radius:10px; padding:12px; margin-bottom:16px;">
    <p>
      The state of two qubits is obtained using the <b>tensor product</b>.
      Suppose the first qubit is
      $|a\rangle = a_1|0\rangle + a_2|1\rangle$,
      and the second qubit is
      $|b\rangle = b_1|0\rangle + b_2|1\rangle$.
    </p>
    <p style="text-align:center; margin:8px 0;">
      \[
      |a\rangle \otimes |b\rangle =
      a_1 b_1 |00\rangle +
      a_1 b_2 |01\rangle +
      a_2 b_1 |10\rangle +
      a_2 b_2 |11\rangle
      \]
    </p>
    <p>
      Here the computational basis is given by:
      \(|00\rangle = [1,0,0,0]^T\),
      \(|01\rangle = [0,1,0,0]^T\),
      \(|10\rangle = [0,0,1,0]^T\),
      \(|11\rangle = [0,0,0,1]^T\).
    </p>
    <p style="margin:10px 0; padding:8px 12px; background:#fef2f2; border-left:4px solid #dc2626; font-size:13px; color:#7f1d1d;">
      ⚡ <b>Remark:</b> We may write the tensor product as
      \(|a\rangle |b\rangle\), or simply \(|ab\rangle\), for convenience.  
      Unlike classical bits (Cartesian product), quantum states combine through
      tensor products, leading to fundamentally richer structures.
    </p>
    </div>
  <!-- </div>
</div> -->


<p style="border: 2px solid #2196F3; border-radius: 8px; padding: 10px; background-color: #E3F2FD; margin:10px 0;">
    <b> ⚡ Tensor Product Rule</b> <br>
    For two column vectors
    $x = \begin{bmatrix}x_1 \\ x_2\end{bmatrix} \in \mathbb{C}^2, \quad
    y = \begin{bmatrix}y_1 \\ y_2\end{bmatrix} \in \mathbb{C}^2,$
    their tensor product is:
      $$
    x \otimes y =
      \begin{bmatrix}
      x_1 \begin{bmatrix} y_1 \\ y_2 \\\end{bmatrix}\\
      x_2 \begin{bmatrix} y_1 \\ y_2 \\\end{bmatrix}
    \end{bmatrix} =
    \begin{bmatrix}
      x_1 y_1 \\
      x_1 y_2 \\
      x_2 y_1 \\
      x_2 y_2
    \end{bmatrix}
    \in \mathbb{C}^4
    $$
  </p>
<div style="background:#ffffff; border:1px solid #cbd5e1; border-radius:10px; box-shadow:0 4px 10px rgba(0,0,0,.08);">
  <div style="padding:10px 14px; font-size:14px; color:#334155;">
    <h4 style="color:green; font-weight:bold; font-size:18px; font-family:'Times New Roman', Times, serif; margin-top:10px;">
        Multi-Qubit States</h4>

  <p>
    Extending beyond two qubits, an <b>N-qubit state</b> (with \(N > 2\)) is represented as a \(2^N\)-dimensional complex vector:
  </p>

  <p style="text-align:center; margin:10px 0;">
    \[
      |\psi\rangle = \sum_{i=1}^{2^N} c_i |i\rangle \;\; \in \mathbb{C}^{2^N},
    \]
  </p>

  <p>
    where the coefficients satisfy the normalization condition
    \(\sum_{i=1}^{2^N} |c_i|^2 = 1\).  
    The computational basis \(|i\rangle\) corresponds to a bit string \(i \in \{0,1\}^N\).
  </p>

  

  <div style="margin:10px 0; padding:8px 12px; background:#f0f9ff; border-left:4px solid #0284c7; font-size:13px; color:#0c4a6e;">
    <strong style="font-size:15px">Computational Basis:</strong>
  <p>
    For an N-qubit system, the computational basis consists of all bit strings:
  </p>

  <p style="text-align:center; margin:10px 0;">
    \[
      \{|0\cdots0\rangle, |0\cdots1\rangle, \ldots, |1\cdots1\rangle\},
    \]
  </p>

  <p>
    These form an <b>orthonormal basis</b> of the Hilbert space \(\mathbb{C}^{2^N}\),  
    satisfying the relation:
  </p>

  <p style="text-align:center; margin:10px 0;">
    \[
      \langle i | j \rangle = \delta_{ij}, \quad \forall i,j \in [2^N].
    \]
  </p>
  </div>

  <h4 style="color:#7e22ce; font-weight:bold; font-size:18px; font-family:'Times New Roman', Times, serif; margin-top:10px;">
        🔗 Entangled Multi-Qubit States</h4>

  <p>
    A key feature of multi-qubit systems is <b>entanglement</b>, a non-classical correlation that cannot be explained by classical physics.
    Entanglement is essential for quantum computational advantage (Jozsa & Linden, 2003).  
    For example, Shor’s algorithm leverages entanglement to achieve exponential speedups.
  </p>

  <p><b>Definition (Entanglement in Dirac Notation):</b></p>

  <p style="margin:10px 0;">
    An N-qubit state \(|\psi\rangle \in \mathbb{C}^{2^N}\) is <b>entangled</b> if it cannot be expressed as:
    $$
      |\psi\rangle \neq |\psi_a\rangle \otimes |\psi_b\rangle,
    $$
    for any partition of the system into subsystems \(A\) and \(B\) with \(N_A + N_B = N\).
  </p>

  <p>
    If such a factorization exists, the state is called <b>separable</b>.  
    In entangled states, the information of one qubit cannot be described independently of the others—even when spatially separated.
  </p>

</div>
</div>
<div style="font-family:sans-serif; line-height:1.55; color:#1e293b; padding:24px; border-left:5px solid #dc2626; background:#fef2f2; border-radius:6px; margin-top:15px;">

  <h3 style="color:#b91c1c; margin-top:5px;">🧮 Density Matrix (Density Operators)</h3>

  <p>
    Another description of quantum states is through the <b>density matrix</b> (or density operator).  
    The need for density operators arises from the fact that real quantum systems are rarely isolated:
  </p>

  <ul style="margin:6px 0 12px 18px;">
    <li><b>Pure states</b> (ideal case): Described by Dirac notation \(|\psi\rangle\).</li>
    <li><b>Mixed states</b> (open system): Described by density operators, when qubits interact with the environment.</li>
  </ul>

  <p>
    Mathematically, for an N-qubit system, the density operator is:
  </p>

  <p style="text-align:center; margin:10px 0;">
    \[
      \rho = \sum_{i=1}^{m} p_i \rho_i, \quad \rho_i = |\psi_i\rangle \langle \psi_i|
    \]
  </p>

  <p>
    where \(|\psi_i\rangle \in \mathbb{C}^{2^N}\), \(p_i \in [0,1]\), and \(\sum_i p_i = 1\).  
    Each \(\rho_i\) is the <b>outer product</b> of a pure state.
  </p>

  <div style="margin:10px 0; padding:8px 12px; background:#fff7ed; border-left:4px solid #d97706; font-size:13px; color:#7c2d12;">
    🔑 <b>Outer Product:</b><br/>
    For vectors \(|u\rangle, |v\rangle \in \mathbb{C}^n\):
    \[
      |u\rangle \langle v| =
      \begin{bmatrix}
        u_1 \\ u_2 \\ \vdots \\ u_n
      \end{bmatrix}
      \begin{bmatrix}
        v_1^* & v_2^* & \cdots & v_n^*
      \end{bmatrix}
    \]
  </div>

  <h4 style="color:#2563eb; margin-top:12px;">Properties of Density Operators</h4>
  <ul style="margin:6px 0 12px 18px;">
    <li>ρ is a <b>positive semi-definite (PSD)</b> matrix: \(0 \preceq \rho\).</li>
    <li>ρ has unit trace: \(\text{Tr}(\rho) = 1\).</li>
    <li>ρ is Hermitian: \(\rho = \rho^\dagger\).</li>
  </ul>

  <div style="margin:10px 0; padding:8px 12px; background:#f0f9ff; border-left:4px solid #0284c7; font-size:13px; color:#0c4a6e;">
    📘 <b>Definition (Positive Semi-Definite Matrix):</b><br/>
    A matrix \(A \in \mathbb{C}^{n \times n}\) is PSD if:
    <ol style="margin:4px 0 0 18px; font-size:13px;">
      <li>Hermitian: \(A = A^\dagger\)</li>
      <li>\(\langle v| A |v \rangle \geq 0, \; \forall |v\rangle \neq 0\)</li>
    </ol>
  </div>

  <h4 style="color:#9333ea; margin-top:12px;">Pure vs Mixed States</h4>
  <ul style="margin:6px 0 12px 18px;">
    <li><b>Pure state:</b> \(m=1 \implies \rho = |\psi\rangle \langle \psi|\).  
      Criterion: \(\text{Tr}(\rho^n) = 1\) for all \(n \in \mathbb{N}^+\).</li>
    <li><b>Mixed state:</b> \(m > 1\), rank(\(\rho\)) > 1.  
      Criterion: \(\text{Tr}(\rho^n) < 1\) for \(n > 1\).</li>
  </ul>

- **Single-Qubit Gates:** X, Y, Z, H, S, T, Rx, Ry, Rz
- **Two-Qubit Gates:** CNOT, CZ, CP, SWAP
- **Three-Qubit Gates:** Toffoli (CCX), CCZ

<h2 style="color:red; font-weight:bold; font-size:24px; font-family:'Times New Roman', Times, serif;">
AerSimulator with Custom Noise
</h2>
<p style="font-family:'Times New Roman', Times, serif; font-size:16px;">
So far, we have been running our quantum circuits on an <b>ideal simulator</b>, meaning that the calculations were
<b>error-free</b>. However, real quantum computers are affected by <b>noise and hardware imperfections</b>.
To simulate realistic quantum hardware, <b>Qiskit Aer</b> allows us to add <b>custom noise models</b> to our circuits.
</p>
<p style="font-family:'Times New Roman', Times, serif; font-size:16px;">
These noise models introduce errors during gate operations or measurements, helping us understand how quantum
algorithms behave on <b>noisy quantum devices</b>.
</p>
<details>
    <summary style= "color:green; font-weight: bold; font-size: 18px; font-family: 'Times New Roman', Times, serif; margin-bottom:10px;">
    Types of Custom Noise in Qiskit Aer
    </summary>
<h4 style="color:#003366; font-weight:bold; font-family:'Times New Roman', Times, serif;">
1. QuantumError (Gate Errors)
</h4>
<p style="font-family:'Times New Roman', Times, serif; font-size:16px;">
These errors occur during <b>quantum gate operations</b>.
</p>
<ul style="font-family:'Times New Roman', Times, serif; font-size:16px;">
<li><b>Depolarizing Error:</b> A commonly used noise model where a qubit randomly changes its state.</li>

<li><b>Thermal Relaxation Error:</b> Simulates decoherence using relaxation times
<i>T<sub>1</sub></i> and <i>T<sub>2</sub></i>.</li>

<li><b>Pauli Error:</b> Introduces specific Pauli operations such as
<b>X, Y, or Z flips</b> with certain probabilities.</li>

<li><b>Reset Error:</b> Represents imperfect resetting of qubits to the
|0⟩ state.</li>

<li><b>Custom Unitary Error:</b> Allows the user to define a custom gate error.</li>
</ul>

<h4 style="color:#003366; font-weight:bold; font-family:'Times New Roman', Times, serif;">
2. ReadoutError (Measurement Errors)
</h4>

<p style="font-family:'Times New Roman', Times, serif; font-size:16px;">
These errors occur during the <b>measurement process</b>.
</p>

<ul style="font-family:'Times New Roman', Times, serif; font-size:16px;">
<li>A qubit in state <b>0</b> may be measured as <b>1</b>.</li>
<li>A qubit in state <b>1</b> may be measured as <b>0</b>.</li>
</ul>

<p style="font-family:'Times New Roman', Times, serif; font-size:16px;">
This models the <b>misclassification errors</b> that occur in real quantum hardware.
</p>

```python
# Circuit
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

noise_model = NoiseModel()

# 1-qubit depolarizing error
error_1q = depolarizing_error(0.01, 1)
# 2-qubit depolarizing error
error_2q = depolarizing_error(0.05, 2)

noise_model.add_all_qubit_quantum_error(error_1q, ['h'])
noise_model.add_all_qubit_quantum_error(error_2q, ['cx'])

# Aer simulator with noise
sim = AerSimulator(
    noise_model=noise_model,
    shots=1024
)

result = sim.run(qc).result()
counts = result.get_counts()

print(counts)
```

```python
plot_histogram(counts, title = "Simulation with Noise")
```

<h2 style="color:red; font-weight:bold; font-size:24px; font-family:'Times New Roman', Times, serif;">
Simulate on Fake Backend
</h2>

<p style="font-family:'Times New Roman', Times, serif; font-size:16px;">
In Qiskit, a <b>FakeBackend</b> is a software model that mimics the behavior of a real
IBM Quantum device. Instead of running a circuit on actual quantum hardware, the
simulation is performed locally on your computer using <b>Qiskit Aer</b>, while still
incorporating the hardware properties and noise characteristics of a real device.
</p>

<p style="font-family:'Times New Roman', Times, serif; font-size:16px;">
This allows researchers and students to test and analyze quantum circuits under
<b>realistic hardware conditions</b> without needing access to an actual quantum computer.
</p>

<details>
    <summary style= "color:green; font-weight: bold; font-size: 18px; font-family: 'Times New Roman', Times, serif; margin-bottom:10px;">
    More on Fake Backend
    </summary>
<h3 style="color:green; font-weight:bold; font-size:20px; font-family:'Times New Roman', Times, serif;">
What Information Does a Fake Backend Include?
</h3>

<p style="font-family:'Times New Roman', Times, serif; font-size:16px;">
A FakeBackend contains many of the same parameters as a real IBM Quantum device:
</p>

<ul style="font-family:'Times New Roman', Times, serif; font-size:16px;">
<li><b>Qubit Connectivity (Coupling Map):</b> Defines which qubits can directly interact
with each other through two-qubit gates such as CNOT.</li>

<li><b>Supported Basis Gates:</b> The native gate set supported by the device
(e.g., <i>sx</i>, <i>x</i>, <i>rz</i>, <i>cx</i>). Circuits must be decomposed into these gates.</li>

<li><b>Gate Errors:</b> Each quantum gate has an associated probability of error
based on real calibration data.</li>

<li><b>Readout Errors:</b> Measurement errors that may cause a qubit in state
<b>|0⟩</b> to be detected as <b>|1⟩</b>, or vice versa.</li>

<li><b>Decoherence Parameters:</b> Realistic <b>T<sub>1</sub></b> (energy relaxation)
and <b>T<sub>2</sub></b> (dephasing) times for each qubit.</li>

<li><b>Gate Duration:</b> The physical time required to apply each quantum gate.</li>

<li><b>Measurement Duration:</b> The time taken to measure qubits on the device.</li>

<li><b>Backend Configuration:</b> Includes number of qubits, device topology, and
available quantum instructions.</li>
</ul>

<h3 style="color:green; font-weight:bold; font-size:20px; font-family:'Times New Roman', Times, serif;">
Why Use Fake Backends?
</h3>

<ul style="font-family:'Times New Roman', Times, serif; font-size:16px;">
<li><b>Hardware-aware testing:</b> Test circuits with realistic noise before running on real quantum hardware.</li>

<li><b>No queue time:</b> Run simulations instantly without waiting in hardware queues.</li>

<li><b>Algorithm benchmarking:</b> Evaluate how noise affects algorithm performance.</li>

<li><b>Education and training:</b> Ideal for teaching quantum computing concepts with realistic device behavior.</li>
</ul>

<h3 style="color:green; font-weight:bold; font-size:20px; font-family:'Times New Roman', Times, serif;">
Example Fake Devices in Qiskit
</h3>

<p style="font-family:'Times New Roman', Times, serif; font-size:16px;">
Qiskit provides several pre-built FakeBackends that emulate real IBM devices, such as:
</p>

<ul style="font-family:'Times New Roman', Times, serif; font-size:16px;">
<li><b>FakeAthens</b></li>
<li><b>FakeManila</b></li>
<li><b>FakeJakarta</b></li>
<li><b>FakeLima</b></li>
<li><b>FakeBelem</b></li>
</ul>

<p style="font-family:'Times New Roman', Times, serif; font-size:16px;">
Each fake backend is built from the calibration data of its corresponding real device,

```python
from qiskit import transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit_ibm_runtime.fake_provider import FakeManilaV2

# Circuit
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

# Fake backend
fake_backend = FakeManilaV2()

# Aer simulator with fake backend noise + config
sim = AerSimulator.from_backend(fake_backend)

# IMPORTANT: transpile for fake backend or Real Quantum Computer
tqc = transpile(qc, fake_backend)

# Run transpiled circuit
result = sim.run(tqc, shots=1024).result()
counts = result.get_counts()

# Plot
plot_histogram(counts, title="Simulation on FakeManila Backend")
```
